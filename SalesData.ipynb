```python
import matplotlib.pyplot as plt
details.groupby("Category")["Profit"].sum().plot(kind="pie",autopct="%1.1f%%",
textprops={"color":"white"})
plt.title("Profit by Category",color="white")
plt.ylabel("")
plt.gca().set_facecolor("black")
plt.gcf().set_facecolor("black")
plt.show
```

```python
details.groupby("Category")["Amount"].sum().plot(kind="bar",
color=["red","green","yellow"],edgecolor="black")
plt.title("Amount By Category",color="white")
plt.xlabel("category",color="white")
plt.ylabel("Amount",color="white")
plt.gca().set_facecolor("black")
plt.gcf().set_facecolor("black")
plt.xticks(color="white")
plt.yticks(color="white")
ax=plt.gca()
ax.spines["left"].set_color("white")
ax.spines["bottom"].set_color("white")
plt.show()
```

```python
merged.groupby("Order Date")["Amount"].sum().plot(kind="line",color="cyan")
plt.title("Total Amount By Date",color="white")
plt.xlabel("Order Date",color="white")
plt.ylabel("Amount",color="white")
plt.xticks(rotation=45)
plt.gca().set_facecolor(color="black")
plt.gcf().set_facecolor(color="black")
plt.xticks(color="white")
plt.yticks(color="white")
ax=plt.gca()
ax.spines["left"].set_color("white")
ax.spines["bottom"].set_color("white")
plt.show()
```

```python
merged.groupby("City")["Profit"].sum().plot(kind="line",color="lime")
plt.title("Total Profit By Cities",color="white")
plt.xlabel("Cities",color="white")
plt.ylabel("Profit",color="white")
plt.xticks(rotation=45)
plt.gca().set_facecolor(color="black")
plt.gcf().set_facecolor(color="black")
plt.xticks(color="white")
plt.yticks(color="white")
ax=plt.gca()
ax.spines["left"].set_color("white")
ax.spines["bottom"].set_color("white")
plt.show()
```

```python
plt.scatter(details["Quantity"],details["Profit"],color="yellow")
plt.title("Quantity vs Profit",color="white")
plt.xlabel("Quantity",color="white")
plt.ylabel("Profit",color="white")
plt.gca().set_facecolor(color="black")
plt.gcf().set_facecolor(color="black")
plt.xticks(color="white")
plt.yticks(color="white")
ax=plt.gca()
ax.spines["left"].set_color("white")
ax.spines["bottom"].set_color("white")
plt.show()
```

```python
details.boxplot(column="Profit",color="cyan")
plt.title("Boxplot of Profit")
```
